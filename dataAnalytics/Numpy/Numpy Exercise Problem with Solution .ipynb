```python
a = np.random.normal(0, 10 , 100)
```

```python
def cal_std(arr):
    mean = np.mean(arr)
    diff = mean - arr
    square =  np.power(diff , 2)
    root = np.power(np.sum(square)/(len(arr)) , 0.5)
    return root
```

```python
cal_std(a)
```
