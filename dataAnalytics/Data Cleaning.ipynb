```python
import pandas as pd
import numpy as np
```

```python
nfl_data = pd.read_csv('NFL Play by Play 2009-2017 (v4).csv')
sf_permits = pd.read_csv('Building_Permits.csv')
```

```python
# get the number of missing data points per column
missing_values_count = nfl_data.isnull().sum()
```

```python
c = missing_values_count.to_frame()
pd.set_option('display.max_rows', None)
c
```

## Total missing values that we have in nfl_data ?

```python
total_cells_nfl = np.product(nfl_data.shape)
total_cells_nfl
```

```python
total_missing_nfl = missing_values_count.sum()
total_missing_nfl
```

```python
# Total % of data i.e, missing
(total_missing_nfl/total_cells_nfl)*100
```

## Total missing values that we have in sf_permits ?


```python
missing_values_count_sf = sf_permits.isnull().sum()
a = missing_values_count_sf.to_frame()
a.head(10)
```

```python
total_cells_sf = np.product(sf_permits.shape)
total_cells_sf
```

```python
total_missing_sf = missing_values_count_sf.sum()
total_missing_sf
```

```python
# Total % of data i.e, missing
(total_missing_sf/total_cells_sf)*100
```

# Drop missing values


```python
# nfl_data.dropna()

# ndl_data Dataset
# remove all columns with at least one missing value
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
columns_with_na_dropped.shape
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % nfl_data.shape[1])
print("Columns in na's dataset: %d \n" % columns_with_na_dropped.shape[1])
```

```python
# sf_permits Dataset
# remove all columns with at least one missing value
columns_with_na_dropped_sf = sf_permits.dropna(axis=1)
columns_with_na_dropped_sf.head()
```

```python
print("Columns in original dataset: %d \n" % sf_permits.shape[1])
print("Columns in na's dataset: %d \n" % columns_with_na_dropped_sf.shape[1])
```

# Filling in missing values automatically

```python
subset_nfl_data = nfl_data.loc[:,'EPA':'Season'].head()
subset_nfl_data
```

```python
# replace all NA's with 0
subset_nfl_data.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the reamining na's with 0
subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```
