```python
country_grp = df.groupby(['Country'])
```

```python
filt = (df['Country'] == 'India')
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_grp['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
def app(x):
    return x.str.contains('Python').sum()
country_grp['LanguageWorkedWith'].apply(app)
# apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['%_Knows_Python'] = (python_df['NumKnowsPython'] / python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='%_Knows_Python', ascending=False, inplace=True)
```
