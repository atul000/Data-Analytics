```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                        columns = np.append(cancer_dataset['feature_names'],['target']))
# c=>containate
cancer_df
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], hue='target')
plt.legend()
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], 
                                hue='target', palette='Dark2')
plt.legend()
```

```python
sns.pairplot(cancer_df, hue='target', x_vars=['mean radius', 'mean texture'], y_vars=['mean radius'])
plt.legend()
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], hue='target',
              kind = 'reg')
plt.legend()
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], hue='target',diag_kind='hist')
plt.legend()
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], hue='target', markers=['*','<'])
plt.legend()
```

```python
sns.pairplot(cancer_df, vars = ['mean radius', 'mean texture', 'mean perimeter', 
                                'mean area', 'mean smoothness'], hue='target', height=20)
plt.legend()
```
