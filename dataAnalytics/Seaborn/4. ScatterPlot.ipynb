```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x=titanic_df.age, y=titanic_df.fare)
```

```python
sns.scatterplot(x = titanic_df['age'], y = titanic_df['fare'])
```

```python
# scatter plot hue parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex")
plt.show()
```

```python
# scatter plot hue_order parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'])
```

```python
# scatter plot size parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who")
```

```python
# scatter plot sizes parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who",
                sizes = (50, 200), hue='who')
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who",
                sizes = (50, 200), hue='who', 
                size_order=['child', 'man', 'woman'])
```

```python
# scatter plot style parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df,
                style = "who", hue='who')
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", 
                style_order=['child','woman','man'], hue='who')
```

```python
# scatter plot palette parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
                palette="gist_ncar") # palette does not work without hue
```

```python
# scatter plot markers parameter
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", 
                markers= ['3','1',3], hue='who')
```

```python
ax = sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue='who')
 
ax.set(xlabel = "Age",
      ylabel = "Fare",
      title = "Seaborn Scatter Plot of Age and Fare")
```

```python
# scatter plot kwrgs (keyword arguments) 
plt.figure(figsize=(16,9)) # figure size in 16:9 ratio
 
kwargs  =   {'edgecolor':"r",
             'facecolor':"k",
             'linewidth':1.7,
             'linestyle':'--',
            }
 
sns.scatterplot(x = "age", y = "fare", data = titanic_df, 
                size = "sex", sizes = (200, 600), 
                alpha = .7,  **kwargs, hue='sex')
```
