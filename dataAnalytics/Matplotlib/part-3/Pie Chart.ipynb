```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]
```

```python
plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%',shadow=True, startangle=95,wedgeprops={'edgecolor': 'black'})
plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```
